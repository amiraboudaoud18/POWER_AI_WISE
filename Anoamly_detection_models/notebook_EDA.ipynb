## ***Anomaly detection part*** 

#### UPS data Anomaly detection

```python
# output Voltage range check
anomalies_voltage_range = pd.DataFrame()

for phase in ['a', 'b', 'c']:
    col = f"ups_v{phase}_out"
    anomaly_rows = df_ups[(df_ups[col] < 200) | (df_ups[col] > 250)]
    anomalies_voltage_range = pd.concat([anomalies_voltage_range, anomaly_rows])

print("Voltage Range Anomalies:", anomalies_voltage_range.shape)
print(anomalies_voltage_range[["ups_va_out", "ups_vb_out", "ups_vc_out"]].head())
```

```python
# Battery behavior
charge_threshold = df_ups["ups_bat_i_charge"].mean() + 3 * df_ups["ups_bat_i_charge"].std()
discharge_threshold = df_ups["ups_bat_i_discharge"].mean() + 3 * df_ups["ups_bat_i_discharge"].std()

anomalies_bat_charge = df_ups[df_ups["ups_bat_i_charge"] > charge_threshold]
anomalies_bat_discharge = df_ups[df_ups["ups_bat_i_discharge"] > discharge_threshold]

print("Battery Charge Anomalies:", anomalies_bat_charge.shape)
print("Battery Discharge Anomalies:", anomalies_bat_discharge.shape)
```

```python
#  Z-score method for voltage 

from scipy.stats import zscore

anomalies_zscore = pd.DataFrame()
for col in ['ups_va_out', 'ups_vb_out', 'ups_vc_out']:
    zscores = zscore(df_ups[col])
    anomalies = df_ups[abs(zscores) > 3]
    anomalies_zscore = pd.concat([anomalies_zscore, anomalies])

print("Z-Score Voltage Anomalies:", anomalies_zscore.shape)

```

```python
# IQR method for ups_ia_out, ups_ib_out, ups_ic_out

anomalies_iqr = pd.DataFrame()
for col in ['ups_ia_out', 'ups_ib_out', 'ups_ic_out']:
    Q1 = df_ups[col].quantile(0.25)
    Q3 = df_ups[col].quantile(0.75)
    IQR = Q3 - Q1
    anomalies = df_ups[(df_ups[col] < Q1 - 1.5 * IQR) | (df_ups[col] > Q3 + 1.5 * IQR)]
    anomalies_iqr = pd.concat([anomalies_iqr, anomalies])

print("IQR Current Anomalies:", anomalies_iqr.shape)

```

```python
# Percentile-based Thresholding 
low = df_ups["ups_va_out"].quantile(0.01)
high = df_ups["ups_va_out"].quantile(0.99)

anomalies_percentile = df_ups[(df_ups["ups_va_out"] < low) | (df_ups["ups_va_out"] > high)]
print("Percentile-Based Anomalies:", anomalies_percentile.shape)
```

```python
# Change Rate Detection
df_ups["ups_va_out_diff"] = df_ups["ups_va_out"].diff().abs()
threshold_diff = df_ups["ups_va_out_diff"].mean() + 3 * df_ups["ups_va_out_diff"].std()

anomalies_diff = df_ups[df_ups["ups_va_out_diff"] > threshold_diff]
print("Voltage Spike/Drop Anomalies:", anomalies_diff.shape)
```

#### PDU data Anomaly detection

```python
# Current Overload Detection - Each PDU has a current limit
for i in range(1, 9):
    col = f"pdu{i}_i"
    threshold = df_pdu[col].mean() + 3 * df_pdu[col].std()
    overload_rows = df_pdu[df_pdu[col] > threshold]
    print(f"\n Overload in {col}: {overload_rows.shape}")

```

```python
#  Power Factor Degradation - A power factor < 0.9 is typically a sign of inefficiency or bad load balance.

for i in range(1, 9):
    col = f"pdu{i}_fp"
    pf_rows = df_pdu[df_pdu[col] < 0.8] # i will see if i will do with 0.9 or 0.8

    print(f"\n Low Power Factor in {col}: {pf_rows.shape}")
    print(pf_rows[[col]].head())
```

```python
# Energy Consistency (kWh Should Not Decrease)- Energy counters (kwh) should only increase or stay the same.
for i in range(1, 9):
    col = f"pdu{i}_kwh"
    energy_drop_rows = df_pdu[df_pdu[col].diff() < 0]

    print(f"\n Energy Drop in {col}: {energy_drop_rows.shape}")
    print(energy_drop_rows[[col]].head())
```

```python
# Z-Score Method (Current)- i may add here for fp facotr - not sure

from scipy.stats import zscore

for i in range(1, 9):
    col = f"pdu{i}_i"
    zscores = zscore(df_pdu[col].fillna(0))
    z_anomalies = df_pdu[abs(zscores) > 3]

    print(f"\n Z-Score Anomalies in {col}: {z_anomalies.shape}")
    print(z_anomalies[[col]].head())
```

```python
# IQR Method
for i in range(1, 9):
    col = f"pdu{i}_i"
    Q1 = df_pdu[col].quantile(0.25)
    Q3 = df_pdu[col].quantile(0.75)
    IQR = Q3 - Q1

    iqr_anomalies = df_pdu[(df_pdu[col] < Q1 - 1.5 * IQR) | (df_pdu[col] > Q3 + 1.5 * IQR)]

    print(f"\n IQR Anomalies in {col}: {iqr_anomalies.shape}")
    print(iqr_anomalies[[col]].head())
```

```python
#  Percentile Method
for i in range(1, 9):
    col = f"pdu{i}_i"
    low = df_pdu[col].quantile(0.01)
    high = df_pdu[col].quantile(0.99)

    pct_anomalies = df_pdu[(df_pdu[col] < low) | (df_pdu[col] > high)]

    print(f"\n Percentile Anomalies in {col}: {pct_anomalies.shape}")
    print(pct_anomalies[[col]].head())
```

```python
# Spike Detection (Change Rate)

for i in range(1, 9):
    col = f"pdu{i}_i"
    diff = df_pdu[col].diff().abs()
    threshold = diff.mean() + 3 * diff.std()

    spike_anomalies = df_pdu[diff > threshold]

    print(f"\n Spikes in {col}: {spike_anomalies.shape}")
    print(spike_anomalies[[col]].head())
```

#### External Power Supply data Anomaly detection

**Domain-Based Sanity Checks (Electrical Rules)**

```python
# Voltage out of safe range (< 200V or > 250V)
for suffix in ['va_1', 'vb_1', 'vc_1', 'va_2', 'vb_2', 'vc_2']:
    col = f"met_{suffix}"
    out_of_range = df_external[(df_external[col] < 200) | (df_external[col] > 250)]
    print(f"\n Voltage anomaly in {col}: {out_of_range.shape}")
    print(out_of_range[[col]].head())
```

```python
#  Power factor degradation (< 0.9)

for suffix in ['fp_1', 'fp_2', 'fpa_1', 'fpb_2', 'fpc_2']:
    col = f"met_{suffix}"
    pf_low = df_external[df_external[col] < 0.9]
    print(f"\n Low Power Factor in {col}: {pf_low.shape}")
    print(pf_low[[col]].head())
```

```python
#  Current imbalance / leakage detection
for col in ['met_i_imb_1', 'met_i_imb_2', 'met_i_leak_1', 'met_i_leak_2']:
    high_vals = df_external[df_external[col] > df_external[col].mean() + 3 * df_external[col].std()]
    print(f"\n High {col}: {high_vals.shape}")
    print(high_vals[[col]].head())
```

```python
# Sequence component anomalies (met_va_seq_n_, met_ia_seq_n_)
# These should be close to 0 in normal balanced conditions.

for col in ['met_va_seq_n_1', 'met_va_seq_n_2', 'met_ia_seq_n_1', 'met_ia_seq_n_2']:
    sequence_abnormal = df_external[df_external[col].abs() > df_external[col].std() * 3]
    print(f"\n Sequence component anomaly in {col}: {sequence_abnormal.shape}")
    print(sequence_abnormal[[col]].head())
```

**Statistical / Probabilistic Anomaly Detection**

```python
# Z-Score Method (for voltage, current, power factor)
from scipy.stats import zscore

for col in ['met_va_1', 'met_ia_1', 'met_fp_1', 'met_va_2', 'met_fp_2']:
    z = zscore(df_external[col].fillna(0))
    anomalies = df_external[abs(z) > 3]
    print(f"\n Z-score anomaly in {col}: {anomalies.shape}")
    print(anomalies[[col]].head())
```

```python
# IQR Method

for col in ['met_va_1', 'met_ia_1', 'met_pa_1', 'met_fp_1']:
    Q1 = df_external[col].quantile(0.25)
    Q3 = df_external[col].quantile(0.75)
    IQR = Q3 - Q1
    iqr_anomalies = df_external[(df_external[col] < Q1 - 1.5 * IQR) | (df_external[col] > Q3 + 1.5 * IQR)]
    print(f"\n IQR anomaly in {col}: {iqr_anomalies.shape}")
    print(iqr_anomalies[[col]].head())
```

```python
# Percentile Check

for col in ['met_va_1', 'met_fp_1']:
    low = df_external[col].quantile(0.01)
    high = df_external[col].quantile(0.99)
    extremes = df_external[(df_external[col] < low) | (df_external[col] > high)]
    print(f"\n Percentile anomaly in {col}: {extremes.shape}")
    print(extremes[[col]].head())
```

#### Energy Consumption Anomaly Detection 

**Domain-Based Checks**

```python
# kWh or kVAh should not decrease

for col in df_energy.columns:
    if "kwh" in col or "kvah" in col:
        decreasing = df_energy[df_energy[col].diff() < 0]
        print(f"\n Energy drop in {col}: {decreasing.shape}")
        # print(decreasing[[col]].head())
```

```python
# Zero energy values (unless expected)
for col in df_energy.columns:
    if df_energy[col].dtype in ['float64', 'int64']:
        zeros = df_energy[df_energy[col] == 0]
        if not zeros.empty:
            print(f"\n Zero values in {col}: {zeros.shape}")
            # print(zeros[[col]].head())
```

**Statistical Checks**

```python
# Z-Score on Monthly or Total Consumption
for col in [c for c in df_energy.columns if "kwh" in c and "month" not in c]:
    z = zscore(df_energy[col].fillna(0))
    anomalies = df_energy[abs(z) > 3]
    print(f"\n Z-score anomaly in {col}: {anomalies.shape}")
    print(anomalies[[col]].head())
```

```python
# Diff/Spike in monthly kWh
for col in df_energy.columns:
    if "month" in col:
        diff = df_energy[col].diff().abs()
        spikes = df_energy[diff > diff.mean() + 3 * diff.std()]
        if not spikes.empty:
            print(f"\n Spike in {col}: {spikes.shape}")
            print(spikes[[col]].head())
```

### TS for UPS data 

```python
# Start and end date
# Convert 'data_hora' to datetime if not already
df['data_hora'] = pd.to_datetime(df['data_hora'])

# 1.1 Start and end date
print("Start:", df['data_hora'].min())
print("End:", df['data_hora'].max())
print("Duration:", df['data_hora'].max() - df['data_hora'].min())

# 1.2 Time frequency check (difference between rows)
df_sorted = df.sort_values('data_hora')
sample_interval = (df_sorted['data_hora'].iloc[1] - df_sorted['data_hora'].iloc[0]).total_seconds()
print("Sample interval (seconds):", sample_interval)

# 1.3 Check for missing timestamps
expected_rows = (df['data_hora'].max() - df['data_hora'].min()).total_seconds() / 10
print("Expected rows:", int(expected_rows))
print("Actual rows:", len(df))
```

```python
# Calculate total number of missing data points (gaps) using 'data_hora' column

# Ensure 'data_hora' is datetime
df_ups['data_hora'] = pd.to_datetime(df_ups['data_hora'])

expected_freq = pd.Timedelta(seconds=10)
df_ups_sorted = df_ups.sort_values('data_hora')
time_diffs = df_ups_sorted['data_hora'].diff()

# Find gaps larger than expected frequency
missing_periods = time_diffs[time_diffs > expected_freq]

# Total number of missing periods (gaps)
print("Number of missing periods:", len(missing_periods))

# Months with missing data
months_with_missing = missing_periods.index.to_series().map(lambda i: df_ups_sorted.iloc[i]['data_hora'].to_period('M')).unique()
print("Months with missing data:", [str(m) for m in months_with_missing])
```
